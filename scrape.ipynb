```python
import requests
import pandas as pd
from bs4 import BeautifulSoup as bs 
import re
```

```python
#Provinsi
provinsi = []
total = []
kd = []  

link = 'https://intranet.atrbpn.go.id/modulWebsite/ajax/daftarPPATProvinsi.php'
resp = requests.get(link)
soup = bs(resp.text, 'html.parser')
tbody = soup.find_all('tbody')
tr = tbody[0].find_all('tr')
for i in range(len(tr)):
    if  tr[i].get('onclick') :
        onclick = tr[i].get('onclick')
        start = onclick.find("('")
        end = onclick.find("')")
        kds = onclick[start + 2 : end]
        kd.append(kds)

        td  = tr[i].find_all('td')

        prov = td[1].text
        provinsi.append(prov)

        jumlah = td[2].text
        total.append(jumlah)
    else:
        break

data = pd.DataFrame(columns=['prov','total','kd'])
data['prov'] = provinsi 
data['total'] = total 
data['kd'] = kd 
data.to_csv('step1.csv', sep='|', index=False)
```

```python
#step1
yd = []
kd_v1 = []
prov_v1 = []
kabupaten = [] 
ppat = []
ppats = []

step1 = pd.read_csv('step1.csv', sep='|')
for i in range(len(step1)):
    prov = step1['prov'][i]
    kd = step1['kd'][i]

    link2 = 'https://intranet.atrbpn.go.id/modulWebsite/ajax/daftarPPATKabupaten.php'
    data = {'kd':kd}
    resp = requests.post(link2, data=data)
    soup = bs(resp.text, 'html.parser')

    tbody = soup.find_all('tbody')
    tr = tbody[0].find_all('tr')
    for x in range(len(tr)-1):
        yds = tr[x].get('onclick')
        start = yds.find("('")
        end = yds.find("')")
        yd.append(yds[start + 2:end])

        kd_v1.append(kd)
        prov_v1.append(prov)

        td = tr[x].find_all('td')
        kabupaten.append(td[1].text)
        ppat.append(td[2].text)
        ppats.append(td[3].text)
        #break
    #break 

data = pd.DataFrame(columns=['prov', 'kabupaten', 'ppat', 'ppats', 'kd', 'yd'])
data['prov'] = prov_v1 
data['kabupaten'] = kabupaten
data['ppat'] = ppat 
data['ppats'] = ppats 
data['kd'] = kd_v1
data['yd'] = yd 
data.to_csv('step2.csv', sep='|', index=False)
data
```

```python
#step3
prov = []
kabupaten = []
ppat = []
ppats = [] 
namas = []
emails = []
sks = []
alamats = []
statuss = []

link = 'https://intranet.atrbpn.go.id/modulWebsite/ajax/daftarPPATDetail.php'
step2 = pd.read_csv('step2.csv', sep='|')
for i in range(len(step2)):
    yd = step2['kd'][i]
    kd = step2['yd'][i] 

    data = {
        'kd' : kd,
        'yd' : yd 
    } 

    resp = requests.post(link, data=data) 
    soup = bs(resp.text, 'html.parser')
    table = soup.find_all('table')
    for x in range(len(table)):
        tr = table[x].find_all('tr')

        prov.append(step2['prov'][i])
        kabupaten.append(step2['kabupaten'][i])
        ppat.append(step2['ppat'][i])
        ppats.append(step2['ppats'][i])

        nama = tr[0].find_all('td')[0].find_all('b')[0].text
        namas.append(nama)

        email = tr[1].find_all('td')[1].text
        emails.append(email) 

        sk = re.sub('\r\n', '', tr[2].find_all('td')[1].text).strip()
        sks.append(sk)

        alamat = tr[3].find_all('td')[1].text
        alamats.append(alamat)

        status = tr[4].find_all('td')[1].text
        statuss.append(status)
        #break
    break 

data = pd.DataFrame(columns=['nama', 'email', 'sk', 'alamat', 'status', 'kabupaten', 'prov', 'ppat', 'ppats'])
data['nama'] = namas
data['email'] = emails 
data['sk'] = sks 
data['alamat'] = alamats 
data['status'] = statuss 
data['kabupaten'] = kabupaten
data['prov'] = prov 
data['ppat'] = ppat 
data['ppats'] = ppats 
data.to_csv('nama_ppat.csv', sep='|', index=False)
```
